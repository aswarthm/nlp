```python
from nltk.corpus import brown
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
# print sentences which has 3 nouns
for sent in brown.sents():
    if(len(sent) > 10):
        tagged_sent = pos_tag(sent)
        nouns = [word for word, pos in tagged_sent if pos == 'NN']
        if len(nouns) >= 3:
            print(nouns)
```
